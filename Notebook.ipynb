### **1\. Top 10 countries with most Covid cases**

```sql
SELECT top 10
      [location]
      ,sum([total_cases]) as [Total_cases]
      --,sum([total_deaths]) as [Total_deaths]
  FROM [COVID].[dbo].[owid-covid-data]
  group by [location]
  order by [location] asc, [total_cases] DESC
```

### **2\. Information on Covid vaccinations based on the continent**

```sql
SELECT
      [continent]
	  , sum([total_vaccinations]) as [Total_vaccinations]
	  , sum([people_vaccinated]) as [People_partially_vaccinated]
	  , sum([people_fully_vaccinated]) as [People_fully_vaccinated]
	  , sum([population]) as [Population]
     FROM [COVID].[dbo].[owid-covid-latest]
	 where [continent] is not null
  group by [continent]
```

### **3\. Analyze people vaccinated in united states per state**

```sql
SELECT
      [location]
      ,sum([total_vaccinations]) as [total_vaccinations]
      ,sum([people_vaccinated]) as [people_vaccinated_partially]
      ,sum([people_fully_vaccinated]) as [people_fully_vaccinated]
  FROM [COVID].[dbo].[USA_state_vaccinations]
  group by [location]
  order by [location] asc
```

### **4\. Post vaccination cases in state of California- Sample State California Data**

```sql
SELECT 
      area,
      sum([unvaccinated_cases]) as [unvaccinated_cases]
      ,sum([vaccinated_cases]) as [vaccinated_cases]
  FROM [COVID].[dbo].[covid-19-post-vaccination-statewide-stats]
  group by area
```

### **5\. Unvaccinated vs Vaccinated cases in the state of California**

```sql
Select area, sum([unvaccinated_hosp]) as [unvaccinated_hosp]
      ,sum([vaccinated_hosp]) as [vaccinated_hosp]
      ,sum([unvaccinated_deaths]) as [unvaccinated_deaths]
      ,sum([vaccinated_deaths]) as [vaccinated_deaths]
       FROM [COVID].[dbo].[covid-19-post-vaccination-statewide-stats]
  group by area
```

### **5\. Vaccination Manufacturers and counts- showing the Vaccine availability across the country**

```sql
SELECT 
      [vaccine],
	  SUM(CONVERT(BIGINT, total_vaccinations)) AS [total_vaccinations]
  FROM [COVID].[dbo].[vaccinations-by-manufacturer]
  group by [vaccine]
  order by [total_vaccinations] desc
```

## **Data cleaning**

### In this below query results we see the covid cases based on the financial status, which was not supposed to be here. We do not encourage our analysis to be done based on the financial status of the population.

```sql
SELECT top 20
      [location]
      ,sum([total_cases]) as [Total_cases]
      ,sum([total_deaths]) as [Total_deaths]
  FROM [COVID].[dbo].[owid-covid-data]
  where location <> 'world'
  group by [location]
  order by [total_cases] DESC
```

### Cleaning the records related to High income, upper middle income and lower middle income. Importing the records into new table before deleting in the source table.

```sql
SELECT [iso_code]
      ,[continent]
      ,[location]
      ,[date]
      ,[total_cases]
      ,[new_cases]
      ,[new_cases_smoothed]
      ,[total_deaths]
      ,[new_deaths]
      ,[new_deaths_smoothed]
      ,[total_cases_per_million]
      ,[new_cases_per_million]
      ,[new_cases_smoothed_per_million]
      ,[total_deaths_per_million]
      ,[new_deaths_per_million]
      ,[new_deaths_smoothed_per_million]
      ,[reproduction_rate]
      ,[icu_patients]
      ,[icu_patients_per_million]
      ,[hosp_patients]
      ,[hosp_patients_per_million]
      ,[weekly_icu_admissions]
      ,[weekly_icu_admissions_per_million]
      ,[weekly_hosp_admissions]
      ,[weekly_hosp_admissions_per_million]
      ,[new_tests]
      ,[total_tests]
      ,[total_tests_per_thousand]
      ,[new_tests_per_thousand]
      ,[new_tests_smoothed]
      ,[new_tests_smoothed_per_thousand]
      ,[positive_rate]
      ,[tests_per_case]
      ,[tests_units]
      ,[total_vaccinations]
      ,[people_vaccinated]
      ,[people_fully_vaccinated]
      ,[total_boosters]
      ,[new_vaccinations]
      ,[new_vaccinations_smoothed]
      ,[total_vaccinations_per_hundred]
      ,[people_vaccinated_per_hundred]
      ,[people_fully_vaccinated_per_hundred]
      ,[total_boosters_per_hundred]
      ,[new_vaccinations_smoothed_per_million]
      ,[new_people_vaccinated_smoothed]
      ,[new_people_vaccinated_smoothed_per_hundred]
      ,[stringency_index]
      ,[population]
      ,[population_density]
      ,[median_age]
      ,[aged_65_older]
      ,[aged_70_older]
      ,[gdp_per_capita]
      ,[extreme_poverty]
      ,[cardiovasc_death_rate]
      ,[diabetes_prevalence]
      ,[female_smokers]
      ,[male_smokers]
      ,[handwashing_facilities]
      ,[hospital_beds_per_thousand]
      ,[life_expectancy]
      ,[human_development_index]
      ,[excess_mortality_cumulative_absolute]
      ,[excess_mortality_cumulative]
      ,[excess_mortality]
      ,[excess_mortality_cumulative_per_million]
	  into Financialstatus
  FROM [COVID].[dbo].[owid-covid-data]
  where location in ('High income','Upper middle income','Lower middle income')
```
